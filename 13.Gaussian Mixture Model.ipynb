#### 예제용 데이터 생성
#### 1. $\mathbf{y} \sim \sum_{k=1}^{K}\pi_{k}\mathcal{N}(\mu_{k},1)$ 을 만족하는 데이터 생성
#### 2. $\pi = (\pi_{1},...,\pi_{k})$ 는 각각 non-negative하면서 합이 1이 되는 값들로 설정
#### 3. 여기에서는 K=3 으로 설정하겠다.

```python
z= np.random.multinomial(1, pvals =[0.2,0.3,0.5],size = 1000)
```

```python
n1 = (z[:,0]==1).sum()
n2 = (z[:,1]==1).sum()
n3 = (z[:,2]==1).sum()
```

```python
print(n1,n2,n3)
```

즉, 예제 데이터는 K=3이다

```python
y = np.concatenate([np.random.normal(-4,1,n1),np.random.normal(0,1,n2),np.random.normal(4,1,n3)])
```

```python
plt.hist(y,bins= 20,color='k', alpha=0.3)
```

#### 1. 데이터를 3개의 Class로 분류를 시도할 것이기 때문에 K=3으로 설정
#### 2. J값은 앞서 예제를 위해 생성한 데이터 개수, iteration 횟수는 임의 지정
#### 3. iteration에서 각 Sample 데이터를 저장할 수 있는 Matrix 생성

```python
K = 3
J = 1000
N = 10000
```

```python
z_mat = np.ones((N,J))
pi_mat = np.zeros((N,K))
mu_mat = np.zeros((N,K))
```

####  1. 우리의 관심 parameter $\Theta = (\mu,\pi) $ 이고 우리는 $p(\mu,\pi\vert y)$ 분포에 관심이 있다.
#### 2. Joint Posterior Distribution으로 바로 Sampling 하는 것은 어려워 각각의 Conditional Posterior Distribution을 구한다.
#### 3. 다만 Sampling을 용이하게 만들기 위해 latent variable $\mathbf{z}$를 사용하기 때문에 우리의 Target Joint Posterior Distribution은 $p(\pi,\mu,z\vert y)$ 이다.
- Sample z from $p(z\vert\mu,\pi,y)$
- Sample $\pi$ from $p(\pi\vert\mu,z,y)$
- Sample $\mu$ from $p(\mu\vert\pi,z,y)$

#### 4. Prior Distribution은 $\pi = (\pi_{1},...,\pi_{k}) \sim \text{Dirichlet}(\frac{1}{K},...,\frac{1}{K}), \quad  \mu_{k} \sim \mathcal{N}(0,10^{2})$

```python
pi_mat[0,:] = [1/3,1/3,1/3]
z_mat[0,:]  = np.random.choice([1,2,3], size=J, replace=True, p=pi_mat[0,:])
mu_mat[0,:] = np.random.normal(0,1,3)
prop = np.zeros(3)
```

```python
for i in range(1,N):        
    for j in range(0,J):
        for k in range(0,K):        
            prop[k] = pi_mat[i-1,k]*np.exp((-0.5)*(y[j]-mu_mat[i-1,k])**2)            
        z_mat[i,j] = np.argmax(np.random.multinomial(1, pvals = prop/prop.sum()))+1
    
    #### Sampling Step for pi
    
    pi_mat[i,:] = np.random.dirichlet([(z_mat[i,:]==1).sum()+1/K,(z_mat[i,:]==2).sum()+1/K,(z_mat[i,:]==3).sum()+1/K],1)
    
    #### Sampling Step for mu & sigma^{2}
    
    for k in range(0,K):
        sig = 1/((z_mat[i,:]==k+1).sum() + 1/(10**2))
        mu = (y[z_mat[i,:]==k+1].sum())*sig     
        mu_mat[i,k] = np.random.normal(mu,np.sqrt(sig),1)     
```

#### 1. 각각의 parameter에 대한 Trace Plot
#### 2. burn-in period 를 iteration 500으로 간주하고 그 이후의 Sampling 데이터만을 가지고 density plot 그리기
#### 3. 각각의 $\mu$ 와 $\pi$는 데이터를 생성할 때 사용했던 $\mu$,$\pi$ 값에 합리적으로 매칭 됨

```python
fig, ax = plt.subplots(1,3, figsize = (15,10))

for iter in range(0,3):
    plt.subplot(2,3,iter+1)
    plt.plot(mu_mat[:,iter],'b')
    plt.xlabel('iteration')
    plt.ylabel('Cluster'+str(iter+1))
    plt.title(r'$\mu$')
```

```python
fig, ax = plt.subplots(1,3, figsize = (15,10))

for iter in range(0,3):
    plt.subplot(2,3,iter+1)
    sns.distplot(mu_mat[501:,iter])
    plt.xlabel('iteration')
    plt.ylabel('Cluster'+str(iter+1))
    plt.title(r'$\mu$')
```

```python
fig, ax = plt.subplots(1,3, figsize = (15,10))

for iter in range(0,3):
    plt.subplot(2,3,iter+1)
    plt.plot(pi_mat[:,iter],'b')
    plt.xlabel('iteration')
    plt.ylabel('Cluster'+str(iter+1))
    plt.title(r'$\pi$')
```
