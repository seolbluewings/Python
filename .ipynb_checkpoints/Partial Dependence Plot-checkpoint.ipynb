```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets.california_housing import fetch_california_housing
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.inspection import partial_dependence
from sklearn.inspection import plot_partial_dependence
```

```python
cal_housing = fetch_california_housing()
```

#### iris 데이터를 pd.DataFrame 형태로 변환

```python
X = pd.DataFrame(cal_housing['data'],columns = cal_housing['feature_names'])
```

```python
X.head()
```

```python
y = pd.DataFrame(cal_housing.target)
```

```python
y.head()
```

#### 간단한 GradientBoost 모델 생성
- 본래 train/test 데이터 구분 후 모델을 생성하는 것이 옳음
- partial dependence plot을 그리기 위함이므로 데이터셋 분리 과정 생략

```python
gbr_model = GradientBoostingRegressor(random_state = 0)
gbr_model.fit(X,y)
```

#### plot을 그리고자 하는 변수들의 idx 값 선택
#### 각 컬럼 명칭 별도의 리스트로 저장

```python
names = cal_housing.feature_names
```

```python
features = [0,1,4,[2,1],[0,2],[3,4]]
```

#### plot_partial_dependence() 함수 옵션
- estimator : 훈련 데이터를 통해 생성한 모델, PDP 그리기 이전에 모델을 생성해야 한다
- X : Input Variables 
- features : PDP를 통해 확인해보고자 하는 X데이터, List 형식으로 입력, 컬럼 idx 번호로 넣어줘야함
- feature_names : str 타입 데이터를 모아놓은 것, j번째 idx에 j번째 컬럼명이 들어가야 정확히 매칭됨
- grid_resolution : 그래프가 꺾이는 수준을 조절하기 위함, 갚을 높일수록 굴곡이 커짐

```python
for iter in range(1,len(features)+1):
    plot_partial_dependence(gbr_model, X, features = [features[iter-1]], feature_names = names)
```
