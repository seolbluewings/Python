## 군집화

### 1.하프문 예시

```python
from sklearn.datasets import make_moons
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.cluster import DBSCAN
```

```python
X, Y = make_moons (n_samples = 1000, noise = 0.05)
```

- 평균 0, 분산 1이 되도록 데이터 스케일 조정

```python
scaler = StandardScaler()
scaler.fit(X)
X_scaled = scaler.transform(X)
```

```python
dbscan = DBSCAN(eps=0.5,min_samples=5)
clusters = dbscan.fit_predict(X_scaled)
```

```python
plt.scatter(X_scaled[:,0],X_scaled[:,1], c= clusters)
plt.xlabel("attribute 1")
plt.ylabel("attribute 2")
```

```python
kmeans = KMeans(n_clusters=2, max_iter = 300000, random_state = 0)
kmeans.fit(X_scaled)
y_predict = kmeans.predict(X_scaled)
```

```python
plt.scatter(X_scaled[:,0],X_scaled[:,1], c= y_predict)
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], marker ='^')
plt.xlabel("attribute 1")
plt.ylabel("attribute 2")
```

### 2.고객 데이터를 바탕으로 고객의 군집을 분류

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"C:\Users\seolbluewings\python\sample\Mall_Customers.csv")
df.head(5)
```

```python
df.rename(index = str, columns =  {'Annual Income (k$)': 'Income',
            'Spending Score (1-100)': 'Score'}, inplace= True)
```

```python
df.head(5)
```

```python
X = df.drop(['CustomerID','Gender'], axis = 1)
```

- 사실상 의미가 없는 고객 CI 번호와 성별을 제외한 나머지 사항 (연령, 소득, 지출점수) 3가지 변수를 가지고 고객의 군집을 결정짓고자 한다.
- 또한, 연령, 소득, 지출점수는 각각 그 크기가 다르기 때문에 3가지 변수에 대해서 모두 평균 0, 분산 1이 되도록 데이터 스케일을 조정할 필요가 있다.

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X)
X_scaled = scaler.transform(X)
```

```python
from sklearn.cluster import KMeans
```

#### 최적의 K값 결정하기 

- K-평균 군집화 방식은 분석자가 설정한 K값에 따라 결과가 달라진다.
- 그래서 최적의 K값을 설정하는 과정이 먼저 이루어지는 것이 타당하다
- 최적의 클러스터를 찾는 방법으로는 엘보(elbow) 기법이 있다.
- 엘보 기법은 클러스터의 개수를 변형시켜 나가면서 SSE 값을 계산하고 SSE 값이 줄어드는 수준이 급격하게 작아지는 부분에서 최적의 클러스터 개수를 결정짓는 방법이다.

```python
clusters = []

for i in range(1,8):
    km = KMeans(n_clusters = i, random_state = 0).fit(X_scaled)
    clusters.append(km.inertia_)

fig, ax = plt.subplots(figsize=(10, 8))
sns.lineplot(x=list(range(1, 8)), y=clusters, marker = 'o', ax=ax)
ax.set_title('Searching for Elbow')
ax.set_xlabel('Numbers of Clusters')
ax.set_ylabel('SSE')
```

- SSE가 줄어드는 수준이 급감하는 부분은 5 또는 6이다. 둘 중 하나의 값을 K값으로 설정하는 것이 좋아보인다.

```python
km5 = KMeans(n_clusters=5, max_iter = 300000, random_state = 0).fit(X_scaled)

X['Labels'] = km5.labels_
plt.figure(figsize=(12, 8))
sns.scatterplot(X['Income'], X['Score'], hue=X['Labels'], palette=sns.color_palette('hls', 5))
plt.title('KMeans with 5 Clusters')
plt.show()
```

```python
km6 = KMeans(n_clusters=6, max_iter = 300000, random_state = 0).fit(X_scaled)

X['Labels'] = km6.labels_
plt.figure(figsize=(12, 8))
sns.scatterplot(X['Income'], X['Score'], hue=X['Labels'], palette=sns.color_palette('hls', 6))
plt.title('KMeans with 6 Clusters')
plt.show()
```

- K값을 설정에 따라 K-평균 군집분석 결과가 다른 것을 확인할 수 있다. 
- 비지도 학습이기 때문에 무엇이 정답이다라고 할 수 없으나 K=6일 때가 더 적절해 보인다. (각자의 주관에 따라 K=5이 더 적절할 수도 있다)
- 적어도 4개 집단에 대해서는 고소득/고지출 부터 저소득/저지출까지 확실히 구분된다.
- 중간소득/중간지출 부분에 대해서는 더 잘 구분할 수 있는 방법에 대해 고민해볼 필요가 있다. 

#### DBSCAN

```python
from sklearn.cluster import DBSCAN
```

- DBSCAN 함수를 활용하는 과정에서 가장 중요한 hyperparameter는 eps 와 min_samples 2가지이다.
- 2가지 hyperparameter를 조정함으로써 더 나은 DBSCAN 모델을 생성해낼 수 있다.

```python
db = DBSCAN(eps=0.45, min_samples=8).fit(X_scaled)

X['Labels'] = db.labels_
plt.figure(figsize=(12, 8))
sns.scatterplot(X['Income'], X['Score'], hue=X['Labels'], 
                palette=sns.color_palette('hls', np.unique(db.labels_).shape[0]))
plt.title('DBSCAN with epsilon 0.5, min samples 6')
plt.show()
```

```python
db = DBSCAN(eps=0.5, min_samples=5).fit(X_scaled)

X['Labels'] = db.labels_
plt.figure(figsize=(12, 8))
sns.scatterplot(X['Income'], X['Score'], hue=X['Labels'], 
                palette=sns.color_palette('hls', np.unique(db.labels_).shape[0]))
plt.title('DBSCAN with epsilon 0.5, min samples 6')
plt.show()
```

- 주어진 데이터에서 DBSCAN은 잘 작동하지 않는다.
- 주어진 데이터의 밀도가 크지 않기 때문에 DBSCAN의 효용성이 떨어진다. 
- Label이 -1로 배정된 것은 Noise로 분류되는 데이터다.
