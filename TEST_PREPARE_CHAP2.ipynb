```python
from sklearn.tree import DecisionTreeClassifier
from sklearn import svm
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
df['Species'].unique()
```

```python
df['N_Species'] = df['Species']
df.loc[(df['Species'] == 'Iris-virginica'),'N_Species'] = 2
df.loc[(df['Species'] == 'Iris-versicolor'),'N_Species'] = 1
df.loc[(df['Species'] == 'Iris-setosa'),'N_Species'] = 0
```

```python
df.head(5)
```

```python
x = df.iloc[:,1:4]
y = df.iloc[:,-1]
y = y.astype('int')
```

```python
svm_model = svm.SVC(kernel = 'rbf')
```

```python
svm_model.fit(x,y)
```

```python
y_hat = svm_model.predict(x)
```

```python
accuracy_score(y,y_hat)
```

```python
clf = DecisionTreeClassifier(random_state = 0)
```

```python
clf.fit(x,y)
```

```python
y_hat = clf.predict(x)
```

```python
accuracy_score(y,y_hat)
```
